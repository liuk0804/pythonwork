## Basic Example

Create a simple `series` object that stores the list of dates as a `DatetimeIndex`.

```python
date_list = [ parse(x) for x in ['2013-1-1', '2013-2-1', '2013-3-1', 
                                 '2013-4-1', '2013-5-1', '2013-6-1' ]]

value_list = np.random.randn(len(date_list))

tmp = pd.Series(value_list, index=date_list)
type(tmp)
print tmp.index
```

```python
tmp.plot()
show()
```

###Resample: upsampling to a higher frequency

```python
tmp = data_by_week.resample('D')
print tmp.head(10)
```

```python
tmp = data_by_week.resample('D',  fill_method='bfill') #Backfill
print tmp.head(10)
```

```python
plot(tmp.index, tmp.values)
plot(data_by_week.index, data_by_week.values, 'o-')
show()
```

```python
print pd.concat([data_by_week, tmp], axis=1).head(10)
```

```python
tmp = data_by_week.resample('D',  fill_method='ffill') 
plot(tmp.index, tmp.values)
plot(data_by_week.index, data_by_week.values, 'o-')
show()
```

```python
print pd.concat([data_by_week, tmp], axis=1).head(10)
```

```python
tmp = data_by_week.resample('D').interpolate()
plot(tmp.index, tmp.values)
plot(data_by_week.index, data_by_week.values, 'o')
show()
```

```python
print pd.concat([data_by_week, tmp], axis=1).head(10)
```

## Moving window functions

```python
mean_10 = pd.rolling_mean(tmp, 10)
```

```python
plot(tmp.index, tmp.values)
plot(data_by_week.index, data_by_week.values, 'o')
plot(mean_10.index, mean_10.values)
show()
```

```python
std_10 = pd.rolling_std(tmp, 10)
plot(tmp.index, tmp.values)
plot(data_by_week.index, data_by_week.values, 'o')
plot(std_10.index, std_10.values)
show()
```
