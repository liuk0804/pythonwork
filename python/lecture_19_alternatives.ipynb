## Rob Story's `vincent`

https://github.com/wrobstory/vincent

```python
import vincent
vincent.core.initialize_notebook()
```

###Bar

```python
group = vincent.GroupedBar(df)
group.axis_titles(x='group', y='Values')
group.legend(title='Types')
group.width=600
group.height=300
group.display()
```

```python
group = vincent.GroupedBar(df.T)
group.legend(title='Categories')
group.colors(brew='Spectral')
group.width=600
group.height=300
group.display()
```

```python
stack = vincent.StackedBar(df)
stack.legend(title='Types')
stack.width=600
stack.height=300
stack.display()
```

```python
stack = vincent.StackedBar(df.T)
stack.legend(title='Types')
stack.width=600
stack.height=300
stack.display()
```

### `ggplot` for python

https://github.com/yhat/ggplot/

```python
from ggplot import *
```

```python
stacked_cars = pd.melt(cars[['mpg','cyl','wt']], id_vars=['cyl','wt'], var_name='quantity')
```

```python
p = ggplot(aes(x='wt', y='value', color='cyl'), data=stacked_cars)
p += geom_point(alpha=0.5, size=100) 
p += xlab("mpg")
p += ylab("weight")
print p
```
